```python
# https://tedboy.github.io/flask/generated/generated/werkzeug.MultiDict.html

from werkzeug.datastructures import MultiDict

args = MultiDict([('author', 'x'), ('author', 'y'), ('sort', 'z')])

print(list(args.keys()))

print(args.getlist('author'))

print(args.getlist('sort'))
```

```python
# https://docs.python-requests.org/en/latest/user/quickstart/#passing-parameters-in-urls
import requests

q = 'Hobbit'
response = requests.get('https://www.googleapis.com/books/v1/volumes', params={'q': q})
response.json()
```
